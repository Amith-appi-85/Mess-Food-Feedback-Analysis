HOSTEL MESS FOOD ANALYZER -

Analyzing student feedback to improve the quality of food in the mess.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

1. What is the overall satisfaction level among students with the mess services?

```python
data = pd.read_csv('mess_feedback_data.csv') #nrows=50 - used to read specific no of rows
df = pd.DataFrame(data)
# print(df.head())

df['avg_rating'] = df['Rating'].mean() #adding new column "avg_rating" to dataset
overall_rating = df['Rating'].mean()
print(f"The Overall Rating Among Students is {overall_rating}")


# The Overall Satisfaction Level Among Students is 3.48 out of 5 , so we can say it is moderately positive experience 

def rating_category(r):
    if r <= 2:
        return 'Poor'
    elif r == 3:
        return 'Average'
    else:
        return 'Good'

df['Satisfaction_Level'] = df['Rating'].apply(rating_category)
df.head()
```

```python
sns.countplot(x = 'Satisfaction_Level' , data = df , order= ['Poor' , 'Average' , 'Good'] , palette= 'rainbow')
plt.title('Distribution of Student Satisfaction Levels')
plt.xlabel('Satisfaction Category')
plt.ylabel('Number of Students')
plt.show()
```

 How do satisfaction levels vary across days of the week?

```python
daywise_rating = df.groupby('Day')['Rating'].mean().reset_index()
print(daywise_rating)

sns.barplot(x ='Day', y = 'Rating' , data = daywise_rating , palette='coolwarm')
plt.title('Daywise Ratings')
plt.xlabel('Days')
plt.ylabel('Ratings')
plt.show()
```

 Which meals (Breakfast or Dinner) receive better ratings overall?


```python
meal_ratings = df.groupby('Meal')['Rating'].mean().reset_index()
print(meal_ratings)

# plt.figure(figsize=(6,2))
sns.barplot(x = 'Meal' , y = 'Rating' , data = meal_ratings , palette='coolwarm')
plt.title('Average Meal Rating Plot')
plt.xlabel('Meal Categories')
plt.ylabel('Average Ratings')
plt.show()

# Lunch Has the Highest Ratings
```

Which day of the week has the lowest & Highest average rating?

```python
sns.barplot(x = 'Day' , y = 'Rating' , data = daywise_rating , palette='dark')
plt.title('Day vs Avg Rating')
plt.xlabel('Days')
plt.ylabel('Average Ratings')


# Monday Has the lowest average ratings
# Friday Has the Highest average ratings

```

How does student satisfaction trend over time?

```python
avg_rating_by_date = df.groupby('Date')['Rating'].mean().reset_index()
print(avg_rating_by_date.head(2))

sns.lineplot(y='Date', x = 'Rating' , data = avg_rating_by_date)
plt.title('Satisfaction Trend Of A Month')
plt.xlabel('Ratings')
plt.ylabel('Dates')
plt.show()


"""

So, looking at our graph of how happy students are with the mess food over the month of June, it's pretty clear: student happiness isn't super steady. It's like a roller coaster , some days they're really loving it, and other days, well, not so much.

There are definitely high-five moments where the ratings shoot up, showing students were super satisfied. But then there are also some "uh-oh" days where the ratings dip quite a bit. This basically tells us that the food quality, or at least how students feel about it, isn't the same every day. It changes a lot!
To make everyone happier in the long run, the folks running the mess should really dig into what made those good days so good, and what went wrong on the not-so-good days. Figuring out those ups and downs is key to making the mess food consistently awesome!"""
```

How are the individual ratings distributed across the entire dataset? Do students generally give high, low, or average ratings?

```python
sns.countplot(x = 'Rating' , data = df , palette='Spectral')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Students')

# generally students have given moderate to high ratings 
```

How do the ratings for Breakfast, Lunch, and Dinner compare to each other on a specific day of the week (e.g., how does Monday Breakfast compare to Monday Lunch and Monday Dinner)?

```python
grouped_rating = df.groupby(['Day' , 'Meal'])['Rating'].mean().reset_index()
print(grouped_rating.head(3)) # to see first 3 rows
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x = 'Day' , y = 'Rating' , hue='Meal' , data = grouped_rating , palette='dark')
plt.title('Average Ratings by Day and Meal Type')
plt.xlabel('Day of The Week')
plt.ylabel('Average Rating')
plt.tight_layout()
```

How many feedbacks were submitted per day? Is there a particular day with more or less engagement? 

```python
pd.to_datetime(df['Date'])

#size() method is used to count total feedbacks in a day
feedback_per_day = df.groupby('Date')['Feedback'].size().reset_index() 
print(feedback_per_day)
```

```python
plt.figure(figsize=(12,6),dpi=100)
sns.barplot(y = 'Date' , x = 'Feedback' , data=feedback_per_day , palette='deep')
plt.title('Feedback Per Day')
plt.xlabel('Date')
plt.ylabel('Number of Feedbacks')
plt.show()


# So , EveryDay 5 Feedbacks Has Been Collected Consistently Throughout a Month of June
```
